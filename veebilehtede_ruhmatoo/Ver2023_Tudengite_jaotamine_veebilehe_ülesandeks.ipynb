# Veebilehe gruppide genereerimine

```python
# Kui lõpus on kolm tudengit alles ning see grupp ei vasta nõuetele, siis lisatakse nendele tudegitele üks grupp juurde ja tehakse 6 tudengi hulgast uus grupp

def uuestiSegamine(grupid, viimased) :

    while True :

          # Selle tsükliga kontrollin, et minu random numbrid oleksid kõik erinevad
          while True:
              tudengidIndeksid = np.random.randint(6, size=(1,3))
              tudengidIndeksid.sort()
              if len(np.unique(tudengidIndeksid[0])) == len(tudengidIndeksid[0]):
                  break

          # Leiame vastavatel indeksitel olevad tudengid
          tudeng1 = viimased[tudengidIndeksid[0][0]]
          tudeng2 = viimased[tudengidIndeksid[0][1]]
          tudeng3 = viimased[tudengidIndeksid[0][2]]

          # Kontrolllime, kas need tudengid saavad olla ühes grupis
          sugu1 = [tudeng1[2], tudeng2[2], tudeng3[2]]
          rahvus1 = [tudeng1[3], tudeng2[3], tudeng3[3]]

          # Kõikide tudengite andmed, et teada saada üle jäänud tudengite grupi andmeid
          sugu2 = [viimased[0][2], viimased[1][2], viimased[2][2], viimased[3][2], viimased[4][2], viimased[5][2]]
          rahvus2 = [viimased[0][3], viimased[1][3], viimased[2][3], viimased[3][3], viimased[4][3], viimased[5][3]]
          if ((sugu1.count("n") != 1) and (rahvus1.count("v") != 1)) and ((sugu2.count("n") - sugu1.count("n") != 1) and (rahvus2.count("v") - rahvus1.count("v") != 1)):

              grupp = []
              eelviimane = [] #Selleks, et moodustatud gruppide juurde lisades oleks uus grupp õige struktuuriga
              # Moodustame nendest ühe grupi
              grupp.append(tudeng1)
              grupp.append(tudeng2)
              grupp.append(tudeng3)

              eelviimane.append(grupp)
              grupid = np.append(grupid, eelviimane, axis = 0)

              # Eemaldame viimaste hulgast
              viimased = np.delete(viimased, tudengidIndeksid[0][2], 0)
              viimased = np.delete(viimased, tudengidIndeksid[0][1], 0)
              viimased = np.delete(viimased, tudengidIndeksid[0][0], 0)

              viimane = [] #Selleks, et moodustatud gruppide juurde lisades oleks uus grupp õige struktuuriga
              viimane.append(viimased)

              # Alles jäänud tudengid lisame moodustatud gruppide hulka.
              grupid = np.append(grupid, viimane, axis = 0)

              return grupid
```
